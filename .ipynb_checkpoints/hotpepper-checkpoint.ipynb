```python
import requests
import json
```

*オブジェクト整形表示関数*

```python
def pretty(obj):
    print(json.dumps(obj, indent=2, ensure_ascii=False))
```

*データ取得関数*

```python
def fetch(target, paramsOption={}):
    baseUrl = 'http://webservice.recruit.co.jp/hotpepper/'
    apiVersion = 'v1'
    defaultParams = {
        'key': 'fe9c64a5d253d6fa',
        'format': 'json'
    }
    url = f'{baseUrl}{target}/{apiVersion}'
    params = dict(defaultParams, **paramsOption)
    response = requests.get(url, params=params)
    return json.loads(response.text)
```

*大エリア（都道府県）データ取得関数*

```python
def getLargeAreas():
    data = fetch('large_area')
    areas = []

    for datum in data['results']['large_area']:
        areas.append({
            'name': datum['name'],
            'code': datum['code']
        })

    return areas
```

*中エリアデータ取得関数*

```python
def getMiddleAreas():
    largeAreas = getLargeAreas()
    middleAreas = []

    for largeArea in largeAreas:
        data = fetch('middle_area', {'large_area': largeArea['code']})

        for datum in data['results']['middle_area']:
            middleAreas.append({
                'name': datum['name'],
                'code': datum['code']
            })

    return middleAreas
```

*小エリアデータ取得関数*

```python
def getSmallAreas():
    middleAreas = getMiddleAreas()
    smallAreas = []

    for middleArea in middleAreas:
        data = fetch('small_area', {'middle_area': middleArea['code']})

        for datum in data['results']['small_area']:
            smallAreas.append({
                'name': datum['name'],
                'code': datum['code'],
                'middle_area': datum['middle_area'],
                'large_area': datum['large_area']
            })

    return smallAreas

# areaList = getSmallAreas()
pretty(len(areaList))
```

### 店舗情報を取得

```python
url = 'http://webservice.recruit.co.jp/hotpepper/gourmet/v1/'
params = {
    'key': 'fe9c64a5d253d6fa',
    'format': 'json',
    'type': 'lite',
    'count': 1,
    'order': 2,  # ジャンルコード順
    'start': 1
}
shopsData = []

for smallArea in smallAreas:
    params['small_area'] = smallArea['code']
    res = requests.get(url, params=params)
    rawData = json.loads(res.text)
    shopsDataData = rawData['results']['middle_area']

    for middleAreaData in shopsDataData:
        shopsData.append({
            'name': middleAreaData['name'],
            'code': middleAreaData['code']
        })

pretty(shopsData)
```

```python
shops = []
for shopData in shopsData:
        shops.append({
        'name': shopData['name'],
        'address': shopData['address'],
        'prefecture': shopData['large_area']['name'],
        'middle_area': shopData['middle_area']['name'],
        'small_area': shopData['small_area']['name'],
        'genre': shopData['genre']['name'],
        'budget': shopData['budget']['name']
    })

pretty(shops)
```
